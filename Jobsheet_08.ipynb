# **Bagian Praktikum**

# Praktikum 1 - Fungsi Dasar dengan 'def'

```python
def sapa():
  print("Halo, Selamat datang di Politeknik Negeri Semarang")

sapa()
```

# Praktikum 2 - Fungsi Parameter dan Return

```python
def tambah(a, b):
  return a + b

hasil = tambah(7, 3)
print("Hasil penjumlahan:", hasil)
```

# Praktikum 3 - Fungsi dengan Default Argument

```python
def sapa(nama = "mahasiswa"):
  print(f"Halo, {nama}!")

sapa()
sapa("Andi")
```

# Praktikum 4 - Fungsi dengan *args dan **kwargs

```python
def jumlahkan(*args):
  return sum(args)

def biodata(**kwargs):
  for key, value in kwargs.items():
    print(f"{key}: {value}")

print("Total:", jumlahkan(1, 2, 3, 4, 5))
biodata(Nama="Budi", Umur=20, Prodi="TRK")
```

# Praktikum 5 - Scope Variabel

```python
x = 10
def ubah():
  x = 5
  print("Nilai Lokal: ", x)

ubah()
print("Nilai Global: ", x)
```

# Praktikum 6 - Fungsi Lambda

```python
kali = lambda x, y: x * y
print("Hasil Perkalian: ", kali(4, 5))
```

# Praktikum 7 - Fungsi Decorator

```python
def decorator_sapa(fungsi):
  def wrapper():
    print("Sebelum fungsi dijalankan....")
    fungsi()
    print("Setelah fungsi dijalankan....")
  return wrapper

@decorator_sapa
def sapa():
  print("Halo, Mahasiswa Polines!")

sapa()
```

# Praktikum 8 - Studi Kasus: Menghitung Rata-rata Nilai Mahasiswa

```python
def hitung_rata(data):
  return sum(data)/len(data)

nilai = [80, 90, 85]
rata = hitung_rata(nilai)

print("Rata-rata nilai: ", rata)
```

# Penugasan 4 - Fungsi dengan Parameter dan Return

```python
def luas_persegi(panjang, lebar):
  return panjang * lebar

print("luas persegi panjang: ", luas_persegi(4, 5))
print("luas persegi panjang: ", luas_persegi(5, 6))
print("luas persegi panjang: ", luas_persegi(10, 9))
```
