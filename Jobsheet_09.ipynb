```python
import math



def luas_lingkaran(r):
  luas = math.pi * (r ** 2)
  return luas

r1 = 7
r2 = 10

print(f"Luas lingkaran dengan r = {r1} yaitu: {luas_lingkaran(r1):.2f}")
print(f"Luas lingkaran dengan r = {r2} yaitu: {luas_lingkaran(r2):.2f}")
```

# Penugasan 6 - Modul Buatan Sendiri

```python
%%writefile konversi.py
def celciusToFahrenheit(c):
  f = (c * 9/5) + 32
  return f

def fahrenheitToCelcius(f):
  c = (f - 32) * 5/9
  return c
```

```python
%%writefile main.py
import konversi

print("=== Program Konversi Suhu ===")

c = 30
f = 100

print(f"{c}ºC = {konversi.celciusToFahrenheit(c):.2f}VF")
print(f"{f}ºF = {konversi.fahrenheitToCelcius(f):.2f}ºC")
```

# Penugasan 8 - Studi Kasus Mini: Game Kuis Matematika

```python
%%writefile laporan.py
def tampilkan_laporan(jumlah_soal, skor):
    print("\n=== Laporan Hasil Kuis ===")
    print(f"Jumlah Soal : {jumlah_soal}")
    print(f"Skor Akhir  : {skor}/{jumlah_soal}")
    if skor == jumlah_soal:
        print("Sempurna! Kamu menjawab semua dengan benar.")
    elif skor >= jumlah_soal / 2:
        print("Bagus! Kamu cukup menguasai materi.")
    else:
        print("Tetap semangat, coba lagi ya!")
```

```python
%%writefile kuis.py
import random
import laporan

jumlah_soal = 5
skor = 0

print("=== Kuis Penjumlahan Sederhana ===")

for i in range(jumlah_soal):
    a = random.randint(1, 10)
    b = random.randint(1, 10)
    print(f"Soal {i+1}: {a} + {b} = ", end="")
    jawaban = int(input())

    if jawaban == a + b:
        print("✅ Benar!\n")
        skor += 1
    else:
        print(f"❌ Salah! Jawaban yang benar adalah {a + b}\n")

laporan.tampilkan_laporan(jumlah_soal, skor)
```
