# **Bagian Praktikum**

# Praktikum 1 - Bubble Sort


```python
def bubble_sort(data):
  n = len(data)
  for i in range (n):
    for j in range (0, n - i - 1):
      if data[j] > data[j+1]:
        data[j], data[j+1] = data[j+1], data[j]
  return data

angka = [64, 34, 25, 12, 22, 11, 90]
print("Data sebelum diurutkan: ", angka)
print("Data setelah bubble sort: ", bubble_sort(angka))
```

# Praktikum 2 - Selection Sort

```python
def selection_sort(data):
  n = len(data)
  for i in range (n):
    min_index = 1
    for j in range(i + 1, n):
      if data[j] < data[min_index]:
        min_index = j
      data[i], data[min_index] = data[min_index], data[i]
    return data

angka = [64, 25, 12, 22, 11]
print("Data sebelum diurutkan: ", angka)
print("Data setelah Selection sort: ", selection_sort(angka))
```

# Praktikum 3 - Insertion Sort

```python
def insertion_sort(data):
  n = len(data)
  for i in range(1, n):
    key = data[i]
    j = i - 1
    while j >= 0 and data[j] > key:
      data[j + 1] = data[j]
      j -= 1
    data[j + 1] = key
  return data

angka = [12, 11,13, 5, 6]
print("Data sebelum diurutkan: ", angka)
print("Data setelah Insertion Sort: ", insertion_sort(angka))
```

# Praktikum 4 - Studi Kasus Nilai Mahasiswa

```python
def bubble_sort_nilai(data):
  n = len(data)
  for i in range(n):
    for j in range (0, n - i - 1):
      if data[j][1] < data[j + 1][1]:
        data[j], data[j + 1] = data[j + 1], data[j]
  return data

mahasiswa = [("Ribut", 70), ("Shifa", 100), ("Raib", 90), ("Ali", 75)]
print("Data sebelum diurutkan: ", mahasiswa)
print("Data setelah diurutkan (desc):", bubble_sort_nilai(mahasiswa))
```

# Praktikum 6 - Sorting Data Barang

```python
def bubble_sort_key(data, key_index=0, reverse=False):
  n = len(data)
  for i in range(n):
    for j in range(0, n - i - 1):
      if reverse:
        if data[j][key_index] < data[j + 1][key_index]:
          data[j], data[j + 1] = data[j + 1], data[j]
      else:
        if data[j][key_index] > data[j + 1][key_index]:
          data[j], data[j + 1] = data[j + 1], data[j]
  return data

barang = [("Buku", 15000), ("Pensil", 5000), ("Tas", 75000), ("Pulpen", 10000)]
print("Urut berdasarkan nama (asc): ", bubble_sort_key(barang.copy(),1))
print("Urut berdasarkan nama (desc): ", bubble_sort_key(barang.copy(), 1, reverse=True))
```

# Penugasan 7 - Analisis Kompleksitas

```python
import time, random

def bubble_sort(data):
  n = len(data)
  for i in range(n):
    for j in range(0, n - i - 1):
      if data[j] > data[j + 1]:
        data[j], data[j + 1] = data[j + 1], data[j]
  return data

def selection_sort(data):
  n = len(data)
  for i in range(n):
    min_index = i
    for j in range (i + 1, n):
      if data[j] < data[min_index]:
        min_index = j
    data[i], data[min_index] = data[min_index], data[i]
  return data

def insertion_sort(data):
  n = len(data)
  for i in range (1, n):
    key = data[i]
    j = i - 1
    while j >= 0 and data[j] > key:
      data[j + 1] = data[j]
      j -= 1
    data[j + 1] = key
  return data


def pengujian(ukuran):
    for i in ukuran:
      data = [random.randint(1,1000) for i in range(i)]
      print(f"\n=== Ukuran data: {i} ===")
      print(data)
      for func in [bubble_sort, selection_sort, insertion_sort]:
          data = data.copy()
          start = time.time()
          end = time.time()
          print(f"{func.__name__} membutuhkan waktu {end - start:.10f} detik")

ukuran = [10, 50, 100]
pengujian(ukuran)
```

# Penugasan 8 - Proyek Mini

```python

def tampilkan_barang(barang):
    print("\nDaftar Barang:")
    print("-" * 30)
    for i, b in enumerate(barang, start=1):
        print(f"{i}. Nama: {b['nama']} | Harga: Rp{b['harga']}")
    print("-" * 30)

jumlah = int(input("Masukkan jumlah barang: "))

daftar_barang = []

for i in range(jumlah):
    print(f"\nBarang ke-{i+1}")
    nama = input("Nama barang  : ")
    harga = int(input("Harga barang : "))

    daftar_barang.append({
        "nama": nama,
        "harga": harga
    })

print("\nPilih metode pengurutan:")
print("1. Urutkan berdasarkan Nama")
print("2. Urutkan berdasarkan Harga")

pilihan_sort = input("Masukkan pilihan (1/2): ")

print("\nPilih urutan:")
print("1. Ascending (A-Z / Murah ke Mahal)")
print("2. Descending (Z-A / Mahal ke Murah)")

urutan = input("Masukkan pilihan (1/2): ")

if pilihan_sort == "1":
    if urutan == "1":
        daftar_barang.sort(key=lambda x: x["nama"].lower())
    else:
        daftar_barang.sort(key=lambda x: x["nama"].lower(), reverse=True)
elif pilihan_sort == "2":
    if urutan == "1":
        daftar_barang.sort(key=lambda x: x["harga"])
    else:
        daftar_barang.sort(key=lambda x: x["harga"], reverse=True)
else:
    print("Pilihan tidak valid!")

tampilkan_barang(daftar_barang)
```
